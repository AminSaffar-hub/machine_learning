# fire up graphlab create

```python
import turicreate
```

# Load some house sales data 

```python
sales = turicreate.SFrame('home_data.sframe')
```

```python
sales
```

# Exploring the data for housing sales 

```python
turicreate.show(sales[1:5000]['sqft_living'],sales[1:5000]['price'])
```

# create a simple regression model of sqft_living to price 

```python
train_data,test_data = sales.random_split(.8,seed=0)
```

# build regression mode

```python
sqft_model = turicreate.linear_regression.create(train_data,target='price',features=['sqft_living'])
```

# Evaluate simple model

```python
print(test_data['price'].mean())
```

```python
sqft_model.evaluate(test_data)
```

# show what predictions look like 

```python
import matplotlib.pyplot as plt 
%matplotlib inline 
```

```python
plt.plot(test_data['sqft_living'],test_data['price'],'.',
         test_data['sqft_living'],sqft_model.predict(test_data),'-')
```

```python
sqft_model.coefficients
```

# explore other features in the data

```python
features = ['bedrooms','bathrooms','sqft_living','sqft_lot','floors','zipcode']
```

```python
sales[features].show()
```

```python
turicreate.show(sales['zipcode'],sales['price'])
```

# build a regression model with more features

```python
my_features_model= turicreate.linear_regression.create(train_data,target='price',features = features)
```

```python
print(sqft_model.evaluate(test_data))
print(my_features_model.evaluate(test_data))
```

# quiz 

```python
advanced_features = [
'bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 'zipcode',
'condition', # condition of house				
'grade', # measure of quality of construction				
'waterfront', # waterfront property				
'view', # type of view				
'sqft_above', # square feet above ground				
'sqft_basement', # square feet in basement				
'yr_built', # the year built				
'yr_renovated', # the year renovated				
'lat', 'long', # the lat-long of the parcel				
'sqft_living15', # average sq.ft. of 15 nearest neighbors 				
'sqft_lot15', # average lot size of 15 nearest neighbors 
]
```

```python
my_advanced_model= turicreate.linear_regression.create(train_data,target='price',features = advanced_features,validation_set=None)
```

```python
print(my_advanced_model.evaluate(test_data))
```

```python
print(my_features_model.evaluate(test_data))
```
